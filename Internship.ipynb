# Finding outliers

```python
data.columns
```

```python
num_cols=['Customer_ID', 'Month', 'Age', 'SSN',
       'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
       'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan',
       'Delay_from_due_date', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit',
       'Num_Credit_Inquiries',  'Outstanding_Debt',
       'Credit_Utilization_Ratio', 'Credit_History_Age', 'Total_EMI_per_month',
       'Amount_invested_monthly', 'Monthly_Balance']
```

```python

for i in num_cols:
    plt.figure()
    plt.boxplot(data[i])
    plt.title(i)
    plt.show()
```

```python
Q1=np.percentile(data['Amount_invested_monthly'],25,interpolation='midpoint')
Q2=np.percentile(data['Amount_invested_monthly'],50,interpolation='midpoint')
Q3=np.percentile(data['Amount_invested_monthly'],75,interpolation='midpoint')
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
up_lim=Q3+1.5*IQR
low_lim=Q1-1.5*IQR
print(up_lim)
print(low_lim)
```

```python
data['Amount_invested_monthly'].median()
```

```python
outliers=[]   #find outliers
for x in data['Amount_invested_monthly']:
    if(x>up_lim)or(x<low_lim):
        outliers.append(x)
outliers
```

```python
ind1=data['Amount_invested_monthly']>up_lim
data.loc[ind1].index
```
