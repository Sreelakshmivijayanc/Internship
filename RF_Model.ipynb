# Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])
```

```python
data
```

# Encoding


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
obj_list = data.select_dtypes(include="object").columns
for obj in obj_list:
    data[obj] = le.fit_transform(data[obj].astype(str))
```

```python
data
```

# Prepare the data

```python

X = data.drop(['Credit_Score','Name','Customer_ID','SSN','Payment_Behaviour','Occupation'], axis=1)
y = data['Credit_Score']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model=lr.fit(X_train,y_train)
y_pred=model.predict(X_test)

```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
logit_reg = LogisticRegression()
logit_reg.fit(X_train,y_train)
```

```python
pred = logit_reg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
print('accuracy =',accuracy_score(y_test,pred))
print('precision =',precision_score(y_test,pred,average='macro'))
print('recall =',recall_score(y_test,pred,average='macro'))
print('f1_score =',f1_score(y_test,pred,average='macro'))

print(confusion_matrix(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,pred)
plt.show()
```

# RandomForestClassifier

```python

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
print('accuracy = ',accuracy_score(y_test, y_pred_rf))
print('precision = ', precision_score(y_test, y_pred_rf, average='macro'))
print('recall = ', recall_score(y_test, y_pred_rf, average='macro'))
print('f1 score = ', f1_score(y_test, y_pred_rf, average='macro'))
confusion_matrix(y_test, y_pred_rf)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,y_pred_rf)
plt.show()
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
pred_cv_tree = tree.predict(X_test)
```

```python
print('accuracy = ',accuracy_score(y_test,pred_cv_tree))
print('precision = ', precision_score(y_test,pred_cv_tree, average='macro'))
print('recall = ', recall_score(y_test,pred_cv_tree, average='macro'))
print('f1 score = ', f1_score(y_test,pred_cv_tree, average='macro'))
confusion_matrix(y_test,pred_cv_tree)
```

```python
confusion_matrix(y_test,pred_cv_tree)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test,pred_cv_tree)
plt.show()
```
